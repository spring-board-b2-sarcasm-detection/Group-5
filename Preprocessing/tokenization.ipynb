```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Load the dataset
data = pd.read_csv("cleaned_dataset.csv")

# Extract the text and labels columns
text = data["Tweet"]
labels = data["Sarcasm"]

# Initialize an empty list to store tokenized tweets and their labels
tokenized_with_labels = []

# Tokenize each tweet and append the tokens with the corresponding label to the list
for tweet, label in zip(text, labels):
    tokens = word_tokenize(tweet)
    tokenized_with_labels.append((tokens, label))

# Display the tokenized array with labels
for tokens, label in tokenized_with_labels[:5]:  # Display the first 5 for brevity
    print(f"Tokens: {tokens}, Label: {label}")
    
```

```python
import gensim.downloader as api
import numpy as np
```

```python
# Load pre-trained GloVe embeddings from gensim
glove = api.load("glove-twitter-25")
```
